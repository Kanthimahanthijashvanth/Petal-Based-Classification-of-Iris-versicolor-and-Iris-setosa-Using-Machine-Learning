```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("train.xlsx")
```

```python
data.head()
```

```python
data.tail()
```

```python
data.isnull()
```

```python
features =data.iloc[:, : -1]
```

```python
features
```

```python
label = data.iloc[:,-1]
```

```python
label
```

```python
model = RandomForestClassifier()
```

```python
model.fit(features,label)
```

```python
testdata = pd.read_excel("test.xlsx")
```

```python
td1 = testdata.iloc[:,:-1]
```

```python
td1
```

```python
prediction = model.predict(td1)
```

```python
prediction
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=testdata,
    x=testdata.columns[0],  
    y=testdata.columns[1],  
    hue='class',  
    palette='Set2',
    s=100
)
plt.title("Predicted Iris Species Classification")
plt.xlabel(testdata.columns[0])
plt.ylabel(testdata.columns[1])
plt.grid(True)
plt.show()
```
